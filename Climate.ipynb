# Средние температуры по месяцам

```python
mean_temp = pd.read_excel('mean_temp.xlsx')
```

```python
#mean_temp.head(10)
```

```python
mean_temp = mean_temp.dropna(axis=0, thresh=4)
mean_temp = mean_temp.dropna(axis=1, thresh=4)
#mean_temp = mean_temp.dropna(axis=0, how='any')
#mean_temp = mean_temp.dropna(axis=1, how='any')
```

```python
head = ['place','jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec','year']
mean_temp.columns = head
mean_temp = mean_temp.drop(index=[1,2])
mean_temp = mean_temp.reset_index(drop=True)
```

```python
mean_temp
```

```python
def place_new(x):
    x = x.replace('*','')
    x = x.replace('1','')
#    x = x.replace('Сургут-Ханты-Мансийский АО','Сургут')
#    x = x.replace('Магадан (Нагаева, бухта)','Магадан')
#    x = x.replace('Уренгой-Ямало-Ненецкий АО','Уренгой')
#    x = x.replace('Хатанга-Таймырский АО','Хатанга')
#    if ' - ' in x:
#        x = x[0:x.find(' - ')]
#        return x
    return x
```

```python
mean_temp['place'] = mean_temp['place'].apply(place_new)
mean_temp['place'] = mean_temp['place'].drop_duplicates(keep = 'first')
```

```python
citylist_mean_temp = mean_temp['place'].unique()
```
